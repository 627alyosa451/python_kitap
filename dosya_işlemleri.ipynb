```python
dosya = "/content/drive/MyDrive/imla_kilavuzu.txt"


def aramaYapma():
  f = open(dosya, "r")

  girdi = input("Aramak istediğiniz kelimeyi giriniz: ")

  for line in f.readlines():
    if line.strip() == girdi:
      print("var")



  f.close()

while True:
  aramaYapma()
  cikis = input("Çıkmak için 'ç'")
  if cikis == "ç":
    break
```
